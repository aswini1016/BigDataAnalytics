```python
from pyspark.sql import SparkSession

# Create a Spark Session
spark = SparkSession.builder \
    .appName("Salary Data Analysis") \
    .getOrCreate()
```

```python
# Load the CSV file into a Spark DataFrame
df = spark.read.csv("Salary_Data.csv", header=True, inferSchema=True)

# Show first 5 rows
df.show(5)
```

```python
# Print schema
df.printSchema()

# Count total rows
print("Total Rows:", df.count())
```

```python
# Summary statistics for numerical columns
df.describe(["Age", "Years of Experience", "Salary"]).show()
```

```python
from pyspark.sql import functions as F

df.groupBy("Gender").agg(
    F.avg("Salary").alias("Avg_Salary"),
    F.max("Salary").alias("Max_Salary"),
    F.min("Salary").alias("Min_Salary")
).show()
```

```python
df.groupBy("Education Level").agg(
    F.round(F.avg("Salary"), 2).alias("Avg_Salary")
).orderBy("Avg_Salary", ascending=False).show()
```

```python
df.groupBy("Years of Experience").agg(
    F.round(F.avg("Salary"), 2).alias("Avg_Salary")
).orderBy("Years of Experience").show()
```

```python
df.groupBy("Job Title").agg(
    F.round(F.avg("Salary"), 2).alias("Avg_Salary")
).orderBy("Avg_Salary", ascending=False).show(5)
```

```python
print("Correlation (Age vs Salary):", df.stat.corr("Age", "Salary"))
print("Correlation (Experience vs Salary):", df.stat.corr("Years of Experience", "Salary"))
```

```python
import matplotlib.pyplot as plt

# Convert to Pandas for visualization
pandas_df = df.toPandas()

# Salary vs Experience Plot
plt.scatter(pandas_df["Years of Experience"], pandas_df["Salary"], alpha=0.6)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Years of Experience")
plt.show()
```

```python
plt.hist(pandas_df["Salary"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.title("Salary Distribution")
plt.show()
```

```python
edu_salary = pandas_df.groupby("Education Level")["Salary"].mean().sort_values(ascending=False)

edu_salary.plot(kind="bar", color="orange", figsize=(8,5))
plt.ylabel("Average Salary")
plt.title("Average Salary by Education Level")
plt.xticks(rotation=45)
plt.show()
```
